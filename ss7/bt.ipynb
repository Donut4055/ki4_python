```python
# bt3
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]

df = pd.DataFrame({'name': name, 'age': age, 'score': score})

print(df.isna())

df['age'].fillna(df['age'].mean().round(1), inplace=True)
print(df)

df.dropna(inplace=True)
print(df)

```

```python
# bt5
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({'name': name, 'age': age, 'score': score})

print(df.groupby('age')['score'].agg([
    ('mean_score', 'mean'),
    ('min_score', 'min'),
    ('max_score', 'max')
]).reset_index())
```

```python
# bt8
import pandas as pd
import glob
import re

all_files = glob.glob('log/*.json')
df_list = []

for file in all_files:
    df_temp = pd.read_json(file, lines = True)
    date_match = re.search(r'log_(\d{4}-\d{2}-\d{2})', file)
    if date_match:
        df_temp['date'] = date_match.group(1)
    df_list.append(df_temp)

df = pd.concat(df_list, ignore_index=True)
df['date'] = pd.to_datetime(df['date'])

daily_report = df.groupby('date')['duration'].sum().reset_index()
daily_report.columns = ['date', 'total_duration']


daily_report.to_csv('daily_report.csv', index=False)

```
