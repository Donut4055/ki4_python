```python
# bt1
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x) 

plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2, color='blue')

plt.title('Biu  u tay ca donut', fontsize=14, fontweight='bold')
plt.xlabel('Trc x', fontsize=12)
plt.ylabel('Trc y', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
# bt4
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 200)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, linewidth=2, color='blue', linestyle='-', label='sin(x)')
plt.plot(x, y_cos, linewidth=2, color='red', linestyle='--', label='cos(x)')

plt.title('So snh sin(x) v cos(x)', fontsize=14, fontweight='bold')

plt.legend(fontsize=12)

plt.show()
```

```python
# bt5
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

y1 = 2 * x + 1
y2 = -x + 10

plt.figure(figsize=(10, 6))
plt.plot(x, y1, linewidth=2, color='blue', label='ng 1')
plt.plot(x, y2, linewidth=2, color='orange', label='ng 2')

plt.title('Giao im ca hai ng thng', fontsize=14, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(loc='upper left', fontsize=12)

plt.show()
```

```python
# bt7
import matplotlib.pyplot as plt
import numpy as np

san_pham = ['o', 'Qun', 'Giy', 'Ti', 'M', 'Tt']
doanh_so = [150, 120, 200, 90, 65, 80]

plt.figure(figsize=(12, 6))
bars = plt.bar(san_pham, doanh_so, color='green', edgecolor='black', linewidth=1.5)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', color='white', fontweight='bold', fontsize=12,
             bbox=dict(boxstyle='round,pad=0.3', facecolor='green', edgecolor='black'))

plt.title('Doanh s bn hng thng 10/2025', fontsize=14, fontweight='bold')

plt.xlabel('Sn phm', fontsize=12)
plt.ylabel('Doanh s (triu)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# bt8
import matplotlib.pyplot as plt
import numpy as np

thang = list(range(1, 13))
nhiet_do = [20, 22, 25, 28, 30, 32, 31, 30, 28, 26, 23, 21]
luong_mua = [10, 15, 30, 120, 200, 250, 280, 220, 150, 80, 30, 15]

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(thang, nhiet_do, color='red', linestyle='-', linewidth=2, marker='o', label='Nhit  (C)')
ax1.set_xlabel('Thng', fontsize=12)
ax1.set_ylabel('Nhit  (C)', fontsize=12, color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.scatter(thang, luong_mua, color='blue', s=80, label='Lng ma (mm)', zorder=5)
ax2.set_ylabel('Lng ma (mm)', fontsize=12, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Nhit  v lng ma H Ni 2025', fontsize=14, fontweight='bold')

ax1.grid(True, linestyle='--', alpha=0.7)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=11)

plt.tight_layout()
plt.show()
```
